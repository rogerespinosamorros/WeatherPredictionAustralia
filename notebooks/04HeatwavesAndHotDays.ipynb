### 🌡️ How Rare Are Heatwaves and Hot Days in Australia?

The **Bureau of Meteorology** defines a “hot day” as one where the **maximum temperature exceeds 35 °C**.

Since 1950, the **frequency of record hot days has more than doubled** — and heatwaves are occurring more often, lasting longer, and starting earlier in the season.

According to the **Climate Council** and **Australian Institute**:
- The **mean annual temperature in Australia has increased by ~1 °C since 1910**
- Cities like Sydney and Darwin are experiencing **significantly more days > 35 °C** compared to previous decades

These changes increase risks of:
- 🔥 Bushfires
- 🧑‍⚕️ Heat-related illness
- ⚡ Energy system failure
- 🚫 Urban uninhabitability

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Total days above 35C, per year
DATA_DIR = r"d:\PersonalProjects\WeatherPredictionAustralia\data"
OUTPUTS_DIR = r"d:\PersonalProjects\WeatherPredictionAustralia\outputs"

heat_df = pd.read_csv(os.path.join(DATA_DIR, "heatDays", "National_Hot_Days__35_C__Simulated_1986_2022_.csv"))
heat_df
```

```python
plt.figure(figsize=(12, 6))
plt.plot(heat_df["Year"], heat_df["Estimated_National_Hot_Days_>35C"], marker='o', color="orangered", linewidth=2)

plt.title("Estimated National Hot Days Over 35 ºC in Australia (1986-2022)")
plt.xlabel("Year")
plt.ylabel("Number of Days > 35 ºC")
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUTS_DIR, "temp", "hot_days_trend.png"), dpi=300)
```
