### 🌊 Floods & Cyclones – Evidence of Increasing Climate Variability

Climate change heightens not only heat and fires, but also **extreme weather events** like floods and cyclones—further threatening Australia's habitability.

#### 🔹 Major Flood Events
- **2022 Eastern Australia Floods** hit Queensland & New South Wales from February to April, causing over **AUD 6.6 billion in damages** and at least 27 fatalities.
- Multiple waterways reached **record river heights**, overwhelming infrastructure and supply chains.

#### 🔹 Tropical Cyclone Frequency
- Australia averages **~11 tropical cyclones per season**, with **4 typically making landfall** 
- Examples:
  - Early 2025 saw **5 tropical lows** near northern Australia—markedly higher than usual.
  - Climate models predict **fewer but more intense cyclones** due to warmer sea surface temperatures.

---

### 🧠 Key Takeaways
- Recent floods are among the **most destructive and costly** in Australian history.
- Cyclone frequency and severity remain high, with **climate change pushing storms toward greater intensity**.
- These events increase natural disaster risks—impacting **urban resilience, infrastructure, and long-term habitability**.

### 🌊 Criteria for Major Flood Events

In this dataset, a **“Major Flood Event”** (`1`) indicates that the year included a nationally significant flood, based on the following criteria:

- 📍 Widespread damage across multiple regions or states
- 💰 Economic impact of **≥ 1 billion AUD**
- ⚠ Official disaster declaration by state or national authorities
- 📰 High public visibility and long-term disruption (e.g., 2011 QLD, 2022 Lismore)

Years without such events are marked as `0`.

> This simplified binary scale helps identify long-term patterns of **climate variability** in Australia.


```python
import pandas as pd
import matplotlib.pyplot as plt
import os

DATA_DIR = r"d:\PersonalProjects\WeatherPredictionAustralia\data"
OUTPUTS_DIR = r"d:\PersonalProjects\WeatherPredictionAustralia\outputs"

```

This dataset summarizes two key indicators of climate variability in Australia:

- **`Major_Flood_Event` (0 or 1)**:  
  A binary value indicating whether a **nationwide-scale or record-breaking flood** occurred in a given year.  
  `1` = A major flood event occurred (e.g., 2011 QLD floods, 2022 Lismore disaster)  
  `0` = No nationally significant flood that year.

- **`Cyclone_Landfalls`**:  
  Estimated number of **tropical cyclones that made landfall** in Australia during that year, based on climatological records and BoM reports.  
  Australia averages ~11 cyclones per season, with ~4–5 typically making landfall.

#### 📚 Sources:
- Bureau of Meteorology (BoM):  
  [Tropical Cyclone Knowledge Centre](https://www.bom.gov.au/cyclone/tropical-cyclone-knowledge-centre/history/climatology/)
- Emergency Management Australia:  
  [AIDR National Disaster Database](https://knowledge.aidr.org.au/resources/australian-disasters/)
- Climate Council & ABC News on 2022 Floods
- Wikipedia:  
  [List of Australian Floods](https://en.wikipedia.org/wiki/List_of_floods_in_Australia)

> These trends show how **climate variability** is intensifying in Australia — with both flooding and cyclone risks rising alongside heat and drought.


```python
import seaborn as sns

flood_and_cyclones_df = pd.read_csv(os.path.join(DATA_DIR, "floods&cyclones", "Flood_Cyclones_Events.csv"))


if "Unnamed: 0" in flood_and_cyclones_df.columns:
    flood_and_cyclones_df = flood_and_cyclones_df.drop(columns=["Unnamed: 0"])
print(flood_and_cyclones_df.columns)

# Set index to "Year" and transpose for heatmap layout
flood_and_cyclones_df = flood_and_cyclones_df.set_index("Year").T

# Heatmap layout
plt.figure(figsize=(14, 6))
sns.heatmap(flood_and_cyclones_df, annot=True, cmap="YlOrBr", cbar=True, linewidths=0.5, linecolor="gray", fmt='g')

# Formatting
plt.title("Heatmap of Flood Events (0/1) and Cyclone Landfalls (2001-2022)")
plt.yticks(rotation=0)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUTS_DIR, "flood_cyclones", "floods_and_cyclones.png"), dpi=300)
plt.show()
```

### 🌡️ Heatmap of Flood and Cyclone Events in Australia (2001–2022)

This heatmap visualizes two key indicators of climate variability:

- **Major Flood Event (Row 1)**  
  Binary values:  
  `1` = A significant nationwide flood occurred (e.g., 2011, 2022)  
  `0` = No major flood recorded that year

- **Cyclone Landfalls (Row 2)**  
  Number of tropical cyclones that made landfall in Australia.  
  Darker shades represent higher cyclone activity.

#### 🧠 Insights:
- Years like **2011 and 2022** show both **major floods and elevated cyclone counts**
- There is **no clear downward trend**, suggesting continued exposure to severe weather events

> This visualization helps identify years of compounding climate risks and supports the broader argument of increasing environmental uninhabitability.
